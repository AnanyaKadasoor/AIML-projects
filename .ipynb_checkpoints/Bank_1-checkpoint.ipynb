```python
import numpy as np
df['Family members'].replace(r'\s+', np.nan, regex=True)
```

```python
df['Family members']=df['Family members'].fillna(1.0)
```

```python
df.describe().transpose()
```

```python
df.columns[df.isnull().any()]
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr=q3-q1
cap1 = q1-1.5*iqr
cap2 = q3+1.5*iqr
df = df[(df>=cap1) & (df<=cap2)]
df = df.dropna(axis=0)
    
print(df.shape)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

cols = df.columns
for i in cols:
    print(i)
    density = gaussian_kde(df[i])
    xs = np.linspace(0,8,200)
    density.covariance_factor = lambda : .25
    density._compute_covariance()
    plt.plot(xs,density(xs))
    plt.show()
```

```python
df1 = df.drop(['ZIP Code','Securities Account','CD Account','ID'],axis=1)
```

```python
df1.shape
```

```python
df1.columns
```

```python
df1['Personal Loan'].value_counts()
```

```python
df1['CreditCard'].value_counts()
```

```python
df1[(df1['Personal Loan']>0) | (df1['CreditCard']>0)].count()
```

```python
df1['ProbOftakingLoan'] = (df1['Personal Loan']>0) | (df1['CreditCard']>0)
```

```python
df1['ProbOftakingLoan'].value_counts()
```

```python
df1['ProbOftakingLoan'] = df1['ProbOftakingLoan']*1
```

```python
df1['ProbOftakingLoan']
```

```python
df1.columns
```

```python
y=df1[['ProbOftakingLoan']]
X=df1.drop(['ProbOftakingLoan'],axis=1)
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set, train_labels, test_labels = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30 , random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
model_score = model.score(X_test, y_test)
print(model_score)
print(metrics.confusion_matrix(y_test, y_predict))
```

```python
model.intercept_
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
print("Report : ", 
   classification_report(y_test, y_predict))
```
