# Q6. Create new data frame with standardize variables and imputation for any missing/outliers - 2 points

```python
from scipy.stats import zscore
numeric_col = car_df.select_dtypes(include = ['int64','float64']).columns
numeric_col
car_df_z = car_df
car_df_z[numeric_col] = car_df[numeric_col].apply(zscore)
car_df_z.head()
```

```python
# Box Plots to check for outliers
fig, axes = plt.subplots(3,4,figsize=(10, 8))
fig.subplots_adjust(left=0.1, bottom=0.1,right=0.99, top=1.5, wspace=0.25, hspace=0.25)
for i, el in enumerate(list(car_df_z.columns.values)[:]):
    a = car_df_z.boxplot(el,ax=axes.flatten()[i])
```

```python
#outliers are observed in acc and hp column
```

```python
#remove the outlier which is above the 3 sigma
car_df_z = car_df_z[car_df_z["hp"] <= 3]
car_df_z.shape
```

```python
car_df_z = car_df_z[car_df_z["acc"] <= 3]
car_df_z.shape
```

# Q7. Create appropriate clusters with the new data set - 2

```python
from sklearn import metrics
cluster_range = range( 2, 10)   # expect 3 to four clusters from the pair panel visual inspection hence restricting from 2 to 6
cluster_errors = []
cluster_sil_scores = []
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 5)
  clusters.fit(car_df_z)
  labels = clusters.labels_                     # capture the cluster lables
  centroids = clusters.cluster_centers_         # capture the centroids
  cluster_errors.append( clusters.inertia_ )    # capture the intertia
  cluster_sil_scores.append(metrics.silhouette_score(car_df_z, labels, metric='euclidean'))
# combine the cluster_range and cluster_errors into a dataframe by combining them
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors , "Avg Sil Score": cluster_sil_scores } )
clusters_df[0:15]
```


# Q8. Identify the appropriate clusters with result from above question - 1.5 points

```python
# Elbow plot

plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

```python
#from cluster 7 onwards the slope change is minimal
```

# Q9. Check for no of values in each cluster and centers for each variables - 1 point

```python
cluster = KMeans( n_clusters = 5, random_state = 1 )
cluster.fit(car_df_z)
```

```python
# Get the centroids.... using function cluster_centers_
centroids = cluster.cluster_centers_
centroids
```

```python
prediction=cluster.predict(car_df_z) # Count of values in each cluster
# Count the values in each cluster
np.bincount(prediction)
```

# Q10. Assign the groups created above to data frame and study the characteristics  for each group. - 2 points

```python
#Append the prediction 
car_df_z["GROUP"] = prediction
car_df_z.groupby("GROUP").count()
```
