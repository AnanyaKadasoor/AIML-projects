###### Outlier Analysis

###### If we try to remove the outliers using IQR logic the personal loan with value "1" count willl be lost.Hence not a good idea to remove the oultliers using this method

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr=q3-q1
cap1 = q1-1.5*iqr
cap2 = q3+1.5*iqr
df = df[(df>=cap1) & (df<=cap2)]
df = df.dropna(axis=0)
    
print(df.shape)
```

### Model the data using Logistic regression

```python
y=df[['Personal Loan']]
X=df.drop(['Personal Loan'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30 , random_state=1)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
model_score = model.score(X_test, y_test)
print(model_score)
```

### Find the accuracy of the model using confusion matrix

```python
#one way to find accuracy using readily avaible functions
```

```python
print("Intercept..",model.intercept_)
print("Accuracy..",accuracy_score(y_test,y_predict))
print("Report : ", 
   classification_report(y_test, y_predict))
print(metrics.confusion_matrix(y_test, y_predict))
```

###### AC=(TN+TP)/(TN+TP+FP+FN) lets caluculate this using above confusion matrix. where TN=True negative, FN=false positive, TP=true positive and FP=false posittive

```python
TN=1340
TP=77
FN=72
FP=11
Accuracy = (TN+TP)/(TN+TP+FP+FN)
print(Accuracy)
```

### Use K - NN model [Hint: Try different values of k] and compare the accuracy of this model with that of Logistic regression 

```python
from sklearn.cross_validation import cross_val_score, cross_val_predict
def knn_model(n):
    NNH= KNeighborsClassifier(n_neighbors= n,metric='minkowski' )
    NNH.fit(X_train,y_train)
    y_pred = NNH.predict(X_test) 
    #y_pred = clf.predict(test_set) 
    print ("n",n,"...Accuracy : ", accuracy_score(y_test,y_pred)*100,"..Test Score...",NNH.score(X_test,y_test),"...Training score...",NNH.score(X_train,y_train))
    return [n,1-NNH.score(X_test, y_test)]
```

```python
x = []
for i in range(1,21,2):
    x.append(knn_model(i))

print(x)
df2 = pd.DataFrame(x)
```
