### Split data in Training and Validation in 80:20 (1 Marks)

```python
### SPLITTING INTO TRAINING AND TEST SETS
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8 , random_state=1)
```

### Do Feature Scaling ( 1 Marks)

```python
# Use StandardScaler or similar methods
from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler().fit(X_train)
rescaledX = scaler_x.transform(X_train)
```

```python
scaler_x_test = StandardScaler().fit(X_test)
rescaledX_test = scaler_x_test.transform(X_test)
```

### Train and Fit NaiveBayes Model(2 Marks)

```python
#Fit the model
model = GaussianNB()
model.fit(rescaledX, y_train)
```

```python
#Predict
model.score(rescaledX , y_train)  
```

### Print Accuracy and Confusion Matrix and Conclude your findings(1 Marks)

```python
# show Confusion Matrix
from sklearn import metrics
from sklearn.metrics import accuracy_score 
test_pred = model.predict(rescaledX_test)

print(metrics.classification_report(y_test, test_pred))
```

```python
# show accuracy
print ("Accuracy : ", 
    accuracy_score(y_test,test_pred)*100)
```

```python
#Show precision and Recall metrics
print(metrics.confusion_matrix(y_test, test_pred))
```
